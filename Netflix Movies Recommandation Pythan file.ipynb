```python
# importing all the Libraries
!pip install googletrans==3.1.0a0
import pandas as pd
#pd.set_option('display.max_columns',None)
#pd.set_option('display.max_rows',None)
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing the Data file as Dataframe 
movies_df=pd.read_csv('netflixData.csv')
movies_df
```

```python
# Modifying the dataframe and keeping only relevent column 
movies_df=movies_df[['Title','Description','Genres','Rating','Content Type']]
movies_df
```

```python
# Dropping the null values
movies_df.dropna()
```

```python
# Cleanin the Title of the movie in the dataframe 
movies_df['Title'] = movies_df['Title'].str.replace(r'[^\w\s]|_', '' ,regex=True)
```

```python
movies_df
```

```python
# Will give Movie REcommandation on the basic sof Genres so conversting the String to Numerical data 
! pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['Genres']).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similar_movies = cosine_similarity(vectors)
```

```python
Similar_movies
```

```python
def recommend(movies):
    movies_index=movies_df[movies_df['Title'] == movies].index[0]
    distance=Similar_movies[movies_index]
    movies_list= sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print (movies_df['Title'].iloc[i[0]])
```

```python
movies_index=movies_df[movies_df['Title'] == 'UnWell'].index[0]
movies_index
distance=Similar_movies[movies_index]
distance
movies_list= sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
movies_list
```

```python
recommend('UnWell')
```
